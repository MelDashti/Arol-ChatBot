```python
from firecrawl import FirecrawlApp

app = FirecrawlApp(api_key="fc-25c10c32cca94744b7b30c838f3a1b53")
```

```python
crawl_status = app.crawl_url(
  'https://www.arol.com/arol-canelli', 
  params={
    'limit': 1, 
    'scrapeOptions': {'formats': ['html'],
    'extract': {
        'prompt': 'Extract the company mission from the page.'
    }}
  },
  poll_interval=30
)
print(crawl_status)
```
